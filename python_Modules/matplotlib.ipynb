# Subplot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)

# PLOT 2
x = np.array([0, 1, 2, 3])  # X-axis values
y = np.array([10, 20, 30, 40])  # Y-axis values

# Create a subplot grid of 1 row and 2 columns, activate the second subplot
plt.subplot(1, 2, 2)

# Plot the data on the second subplot
plt.plot(x, y)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example dataset: Ages of 500 individuals (simulated data)
np.random.seed(42)  # For reproducibility
ages = np.random.randint(18, 80, size=500)  # Generate random ages between 18 and 80

# Create the histogram
plt.hist(ages, bins=10, color='blue', edgecolor='black', alpha=0.7)

# Add titles and labels
plt.title("Age Distribution of Participants")
plt.xlabel("Age Groups")
plt.ylabel("Number of Individuals")

# Show gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the histogram
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate an arbitrary dataset
np.random.seed(42)  # For reproducibility

# Scatter Plot: Hours studied vs Exam scores
hours_studied = np.random.uniform(1, 10, 50)  # Random hours between 1 and 10
exam_scores = hours_studied * 10 + np.random.normal(0, 5, 50)  # Linear relation with noise

# Bar Chart: Grade categories
grades = ['A', 'B', 'C', 'D', 'F']
grade_counts = [12, 20, 10, 5, 3]  # Arbitrary counts of students in each grade

# Histogram: Age distribution of students
ages = np.random.randint(15, 25, 100)  # Random ages between 15 and 25

# Plotting

# 1. Scatter Plot
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)  # First plot in a 1x3 grid
plt.scatter(hours_studied, exam_scores, color='blue', alpha=0.7, edgecolor='black')
plt.title("Hours Studied vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Scores")
plt.grid(True, linestyle='--', alpha=0.5)

# 2. Bar Chart
plt.subplot(1, 3, 2)  # Second plot in a 1x3 grid
plt.bar(grades, grade_counts, color='green', alpha=0.7, edgecolor='black')
plt.title("Grade Distribution")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 3. Histogram
plt.subplot(1, 3, 3)  # Third plot in a 1x3 grid
plt.hist(ages, bins=5, color='purple', alpha=0.7, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the combined plots
plt.tight_layout()
plt.show()
```
