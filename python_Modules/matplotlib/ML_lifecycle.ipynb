```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define the stages of the ML project life cycle
stages = [
    "Problem Definition", "Data Collection", "Data Preparation",
    "Exploratory Data Analysis", "Model Selection", "Model Training",
    "Model Evaluation", "Model Deployment", "Monitoring and Maintenance"
]

# Add nodes to the graph
G.add_nodes_from(stages)

# Define edges (connections between the stages)
edges = [
    ("Problem Definition", "Data Collection"),
    ("Data Collection", "Data Preparation"),
    ("Data Preparation", "Exploratory Data Analysis"),
    ("Exploratory Data Analysis", "Model Selection"),
    ("Model Selection", "Model Training"),
    ("Model Training", "Model Evaluation"),
    ("Model Evaluation", "Model Deployment"),
    ("Model Deployment", "Monitoring and Maintenance")
]

# Add edges to the graph
G.add_edges_from(edges)

# Define the layout for the nodes
pos = nx.spring_layout(G, seed=42)

# Draw the nodes with labels
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='lightblue', edgecolors='black')
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')

# Draw the edges
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray')

# Title and display the diagram
plt.title("Machine Learning Project Life Cycle", fontsize=14)
plt.axis('off')
plt.show()
```
