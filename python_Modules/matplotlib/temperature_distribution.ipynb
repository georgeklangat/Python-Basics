```python
import matplotlib.pyplot as plt
import numpy as np

# weather data for week 1

week_1 = np.array([2,34,5,7,4,56,34,67])
week_2 = np.array([23,45,67,34,67,46,45,78])

plt.plot(week_1, marker='o', color='red',label = 'week_1')
plt.plot(week_2, marker='x', color='blue',label = 'week_2')
plt.xlabel('Day')
plt.ylabel('Temperatures')
plt.title('Temperature Distribution')
plt.legend()
plt.show()
```
